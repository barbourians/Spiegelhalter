### Python libraries:

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Matplotlib settings:

```python
%matplotlib inline
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
plt.style.use('seaborn-whitegrid')
```

### Open the csv file:

```python
jb1 = pd.read_csv('jelly-beans.csv')
```

### Display the data:

```python
jb1
```

```python
jb1.head()
```

```python
jb1.tail()
```

```python
jb1.shape
```

```python
jb1.columns
```

```python
jb1['guess']
```

### Plot the data:

```python
# Plot
plt.plot(jb1)
plt.show()
```

```python
# Dot plot
plt.plot(jb1, 'bo')
plt.show()
```

```python
# Increase the size of the graph
plt.figure(figsize=(20,10))
plt.plot(jb1, 'bo')
plt.show()
```

```python
# Sorting the data may help
jbs = jb1.sort_values(by=['guess'], ignore_index=True)

plt.figure(figsize=(20,8))
plt.plot(jbs, 'bo')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(6,8))
plt.boxplot(jb1)
plt.show()
```

```python
# Use help if you are not sure what a graph shows
?plt.boxplot
```

```python
# Histogram
plt.hist(jb1)
plt.show()
```

```python
# Histogram with more 'bins'
plt.hist(jb1, bins=30)
plt.show()
```

```python
# Histogram with title and labels
plt.title('Guesses of number of jelly beans in the jar')
plt.xlabel('Number Guessed')
plt.ylabel('Count')
plt.hist(jb1, bins=50, color='blue', edgecolor='black')
plt.show()
```

### Describe the data:

```python
jb1.describe()
```

### We could filter out the outliers:

```python
# guess >= 400 and guess < 9000
jb2 = jb1.loc[(jb1['guess'] >= 400) & (jb1['guess'] < 9000)]
```

```python
# Dot plot
plt.figure(figsize=(20,10))
plt.plot(jb2, 'bo')
plt.show()
```

```python
# Box plot - with filtered data
plt.figure(figsize=(6,8))
plt.boxplot(jb2)
plt.show()
```

```python
# Difference in mean values
print(jb1.mean())
print(jb2.mean())
```

```python
plt.hist(jb2, bins=20, color='blue', edgecolor='black')
plt.show()
```

```python
jb2.describe()
```

### Or we could convert the data to a logorithmic scale:

```python
jb3 = np.log10(jb1)
```

```python
jb3.head()
```

```python
# Dot plot
plt.figure(figsize=(20,8))
plt.plot(jb3, 'bo')
plt.show()
```

```python
# 10 to the power of 3.5
pow(10,3.5)
```

```python
# Bell curve
ax = jb3.plot.kde()
plt.show()
```
