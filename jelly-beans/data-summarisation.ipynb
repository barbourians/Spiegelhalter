### We could filter out the outliers:

```python
# guess >= 400 and guess < 9000
jb2 = jb1.loc[(jb1['guess'] >= 0) & (jb1['guess'] < 5000)]
```

```python
# Dot plot
plt.plot(jb2, 'bo')
plt.show()
```

```python
plt.boxplot(jb2)
plt.show()
```

```python
jb2.describe()
```

```python
plt.hist(jb2, bins=20)
plt.show()
```
