### Plot the data:

```python
# Plot
plt.plot(jb1)
plt.show()
```

```python
# Dot plot
plt.plot(jb1, 'bo')
plt.show()
```

```python
# Increase the size of the graph
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(jb1, 'bo')
plt.show()
```

```python
# Set plot size back to the defaults
plt.rcParams["figure.figsize"] = (6,4)
```

```python
# Box plot
plt.boxplot(jb1)
plt.show()
```

```python
# Histogram
plt.hist(jb1)
plt.show()
```

```python
# Histogram with more 'bins'
plt.hist(jb1, bins=30)
plt.show()
```

```python
# Histogram with title and labels
plt.title('Guesses of number of jelly beans in the jar')
plt.xlabel('Number Guessed')
plt.ylabel('Count')
plt.hist(jb1, bins=50, color='blue', edgecolor='black')
plt.show()
```

### Describe the data:

```python
jb1.describe()
```

### We could filter out the outliers:

```python
# guess >= 400 and guess < 9000
jb2 = jb1.loc[(jb1['guess'] >= 400) & (jb1['guess'] < 9000)]
```

```python
# Dot plot
plt.rcParams["figure.figsize"] = (20,8)
plt.plot(jb2, 'bo')
plt.show()
```

```python
# Set plot size back to the defaults
plt.rcParams["figure.figsize"] = (6,4)
```

```python
plt.boxplot(jb2)
plt.show()
```

```python
# Difference in mean values
print(jb1.mean())
print(jb2.mean())
```

```python
plt.hist(jb2, bins=20, color='blue', edgecolor='black')
plt.show()
```

```python
jb2.describe()
```
